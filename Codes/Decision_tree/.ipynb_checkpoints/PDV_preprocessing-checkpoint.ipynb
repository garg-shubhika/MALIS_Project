```python
# Data wrangling
import pandas as pd
import numpy as np

# Data visualisation
import seaborn as sns
import matplotlib.pyplot as plt

# Machine learning
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.datasets import load_digits
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, GridSearchCV 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../../data/Clean_Dataset.csv').drop(['flight','Unnamed: 0'], axis=1)
data.head()
```

```python
print("Data shape: ", data.shape)
```

```python
for i in data.columns.drop(['source_city']).drop(['destination_city']).drop(['arrival_time']).drop(['price']):
    plt.figure(figsize = (10, 2))
    sns.barplot(x = i, y='price', data = data)
```

```python
plt.figure(figsize = (7, 5))
sns.distplot(data['price'], label = 'Skewness: {:.2f}'.format(data['price'].skew()))
plt.legend(loc = 'best')
plt.title('Distribution of price')
```

```python
ohe = OneHotEncoder(sparse = False)
```

```python
list(data['stops'].unique())
```

```python
time = ['Early_Morning', 'Morning', 'Afternoon', "Evening", "Night", "Late_Night"]
oe_time = OrdinalEncoder(categories = [time])

escales = list(data['stops'].unique())
oe_escales = OrdinalEncoder(categories = [escales])
```

```python
X = data.drop('price', axis = 1)
Y = data['price']
```

```python
column_transform = make_column_transformer(
    (ohe, ['airline', 'source_city', 'destination_city', 'class']), 
    (oe_time, ['departure_time']),
    (oe_escales, ['stops']),
    (oe_time, ['arrival_time']))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", Y_train.shape)
print("X_test shape: ", X_test.shape)
print("Y_test shape: ", Y_test.shape)
```

```python
X_useless, X_fast, Y_useless, Y_fast = train_test_split(X, Y, test_size = 0.1, random_state = 42)
print("X_fast shape: ", X_fast.shape)
print("Y_fast shape: ", Y_fast.shape)
```

```python
lm = DecisionTreeRegressor()
lm_pipeline = make_pipeline(column_transform, lm)
lm_pipeline
```

```python
param_grid = {
    'decisiontreeregressor__criterion' : ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    'decisiontreeregressor__max_depth' : [ 10, 14 , 20 ]}
#'decisiontreeregressor__splitter' : ["best", "random"],
search = GridSearchCV(lm_pipeline, param_grid, refit='neg_mean_squared_error')
search.fit(X_fast, Y_fast)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```

```python
lm_pipeline.fit(X_train, Y_train)
lm_predictions = lm_pipeline.predict(X_test)
print("First 5 LM predictions: ", list(lm_predictions[:5]))
```

```python
lm_mae = mean_absolute_error(lm_predictions, Y_test)
lm_rmse =  np.sqrt(mean_squared_error(lm_predictions, Y_test))
print("LM MAE: {:.2f}".format(round(lm_mae, 2)))
print("LM RMSE: {:.2f}".format(round(lm_rmse, 2)))
```
