# Flight Prices

```python
import numpy as np
import pandas as pd
from sklearn import linear_model

df = pd.read_csv('C:/git/MALIS_Project/data/Clean_Dataset.csv')
df2 = df.rename(columns={'class':'ticket_class'})
df2
```

```python
# creating a new dataframe, keeping only the columns that matter

newdf = df2.drop(columns=['flight', 'Unnamed: 0'], axis = ('columns'))
newdf
```

## Creating the dummy variables

```python
airline_D = pd.get_dummies(newdf.airline)
airline_D
```

```python
source_city_D = pd.get_dummies(newdf.source_city)
source_city_D
```

```python
departure_time_D = pd.get_dummies(newdf.departure_time)
departure_time_D
```

```python
stops_D = pd.get_dummies(newdf.stops)
stops_D
```

```python
arrival_time_D = pd.get_dummies(newdf.arrival_time)
arrival_time_D
```

```python
destination_city_D = pd.get_dummies(newdf.destination_city)
destination_city_D
```

```python
class_D = pd.get_dummies(newdf.ticket_class)
class_D
```

```python
merged = pd.concat([newdf, airline_D, source_city_D, departure_time_D, stops_D, arrival_time_D, destination_city_D, class_D], axis=('columns'))
final = merged.drop(['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'ticket_class'], axis='columns')
final
```

```python
y = final.price
x = final.drop(['price'], axis='columns')
```

```python
my_model = linear_model.LinearRegression()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,train_size=0.8)

my_model.fit(X_train, y_train)
accuracy = my_model.score(X_test, y_test)*100
print("My model accuracy is", accuracy, '%')
```
